```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x,y):
    return (x**2 + y**2 - 1)**3 - x**2 * y**3
```

```python
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)

X,Y = np.meshgrid(x,y)

Z= f(X,Y)
```

```python
plt.contour(X, Y, Z, levels=[0], colors='r')
plt.show()
```

```python
x = np.linspace(-1.5, 1.5, 100)
y = np.linspace(-1.5, 1.5, 100)
z = np.linspace(-5, 5, 100)

X, Y, Z = np.meshgrid(x, y, z)

heart_equation = (X**2 + (9/4)*Y**2 + Z**2 - 1)**3 - X**2*Z**3 - (9/200)*Y**2*Z**3

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[heart_equation < 1e-2000], Y[heart_equation < 1e-2000], Z[heart_equation < 1e-2000], color='r', s=3)

plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# --- Equation for L
x1 = np.linspace(0.1, 10, 400) 
y1 = 1 / x1
axes[0].plot(x1, y1, color='r')
axes[0].set_ylim(0, 10)

# --- Equation for O
x2 = np.linspace(-3, 3, 400)
y_positive = np.sqrt(9 - x2**2)
y_negative = -np.sqrt(9 - x2**2)
axes[1].plot(x2, y_positive, color='r')  
axes[1].plot(x2, y_negative, color='r') 

# --- Equation for V
x3 = np.linspace(-3, 3, 400)
y3 = np.abs(-2 * x3)
axes[2].plot(x3, y3, color='r')


# --- Equation for E
y4 = np.linspace(-np.pi, np.pi, 400)
x4 = -3 * np.abs(np.sin(y4))
axes[3].plot(x4, y4, color='r')


plt.tight_layout()
plt.show()
```
